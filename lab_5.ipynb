Create a scatter plot for the following 2 lists showing the correlation between average years of experience and salary.

Years of experience = [1, 2, 5, 10, 15]

Salary = [45000, 60000, 72000, 85000, 98000]

```python
import matplotlib.pyplot as plt
```

```python
# given data
years_exp = [1, 2, 5, 10, 15]
salary = [45000, 60000, 72000, 85000, 98000]
```

```python
# create scatter plot
plt.scatter(years_exp, salary, marker='o', color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.grid(True)
plt.show()
```
