Instructions:

Edit Lab-assignment 1 to check if score inputted for each of the 3 test scores is greater than zero. If score inputted is greater than zero, then consider that score when computing the sum and average of the scores. If any score is less than zero, then do not use that score in the computation of the total and average.

Print the total and average scores.

```python
# Lab 2

grades = []
entries = 0

while entries < 3:
    print(f"Entry {entries + 1} of 3")
    user_input = input("Enter test score (or 'done' to finish): ")

    if user_input.lower() == 'done':
        break

    try:
        score = float(user_input)
        entries += 1
        if score > 0:
            grades.append(score)
        else:
            print("Score is not greater than zero; it will be excluded.")
    except ValueError:
        print("Invalid input. Please enter a number.")


if grades:
    total = sum(grades)
    average = total / len(grades)
    print(f"Total: {total:.2f}")
    print(f"Average: {average:.2f}")
else:
    print("No grades entered.")
```
