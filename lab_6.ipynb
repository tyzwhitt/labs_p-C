You can create multiple charts within one figure by using the subplots=True argument. This creates separate charts for each column of the DataFrame.

Example:
# Creating subplots for Sales and Profit

df.plot(x='Month', y=['Sales', 'Profit'], kind='line', subplots=True, layout=(2, 1), figsize=(8, 6))

# Show the chart

plt.show()

This will create two line charts, one for Sales and one for Profit, stacked vertically.

Key Pandas Plot Types
Here’s a quick reference for different plot types available in Pandas:

- Line chart — `kind='line'`
- Bar chart (vertical) — `kind='bar'`
- Bar chart (horizontal) — `kind='barh'`
- Histogram — `kind='hist'`
- Pie chart — `kind='pie'`
- Scatter plot — `kind='scatter'`
- Box plot — `kind='box'`

```python
# Create a plot to compare the stock price for Jan and Feb
# create a column list

column_list1 = ['Jan', 'Feb']

df[column_list1].plot()

# add a title to chart [Monthly stock price]

plt.title('Monthly Stock Price')

# Clean up X-axis

tick_val = []

tick_lab = []

plt.xticks(tick_val, tick_lab)

# add label to y-axis [Price in US Dollars]

plt.ylabel('Price in US Dollars')

plt.show()
```
