Define a function that would take 2 scores as arguments. 
Compute the average of these 2 scores inside the function.
Your function should return the average value.
Write a Python program to call this function.
Once the value of average score is returned to the main function, print this value.

```python
# create a function to calculate the average of two scores
def calculate_average(score1, score2):
    average = (score1 + score2) / 2
    return average


# Call the function and print the result
avg_score = calculate_average(85, 90)
print("The average score is:", avg_score)
```
